```python
X = wine_data.drop('quality', axis = 1)
```

```python
X
```

```python
y = wine_data['quality'].apply(lambda y_value : 1 if y_value >=7 else 0)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
y.shape, y_train.shape, y_test.shape
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
X_test_pred = model.predict(X_test)
```

```python
model_pipepline = Pipeline(steps=[
    ('model', RandomForestClassifier(criterion = 'entropy', max_depth=4))
])
```

```python
model_pipepline.fit(X,y)
```

```python
pickle.dump(model_pipepline, open("wine-quality-prediction.pkl", "wb"))
```

```python
accuracy_score = accuracy_score(X_test_pred, y_test)
```

```python
print("Accuracy Score:\n", accuracy_score)
```
